# Data cleaning

```python
#convert the original excel downloaded from the website to csv
import pandas as pd

def xlsx_to_csv_pd():
    data_xls1 = pd.read_excel('Tourism_Statistics_Market_2016M10-2019M10.xlsx', index_col=0)
    data_xls1.to_csv('Tourism_Statistics_Market_2016M10-2019M10.csv', encoding='utf-8')
    
    data_xls1 = pd.read_excel('Tourism_Statistics_Total_2016M10-2019M10.xlsx', index_col=0)
    data_xls1.to_csv('Tourism_Statistics_Total_2016M10-2019M10.csv', encoding='utf-8')
    
    data_xls1 = pd.read_excel('Tourism_Statistics_Region_2016M10-2019M10.xlsx', index_col=0)
    data_xls1.to_csv('Tourism_Statistics_Region_2016M10-2019M10.csv', encoding='utf-8')

if __name__ == '__main__':
    xlsx_to_csv_pd()
```

```python
df1 = pd.read_csv('Tourism_Statistics_Market_2016M10-2019M10.csv') #open the csv
df1.head()
```

```python
df1['date'] = df1['year'].map(str) + '-' + df1['month'].map(str) # creat the date format
df1['date'] = pd.to_datetime(df1['date']) # transform the str to datetime
df1['total arrivals'] = df1.loc[:, 'total arrivals'].apply(lambda x: float(x.replace(",", ""))) 
df1['total arrivals'].replace(",", "") # delete the ',' in the colomns'total arrivals'
df1.head()
```

```python
df2 = df1[['market regions / markets','date','total arrivals','year on year (yoy) growths (%)']] # select the columns needed
df2.columns = ['markets','date','total arrivals','year on year (yoy) growths (%)'] # rename the name of the columns
df2.head()
```

```python
df2.to_csv('Tourism_Statistics_Market_2016M10-2019M10(cleaned).csv',index = False) # output the csv
```

```python
df3 = pd.read_csv('Tourism_Statistics_Region_2016M10-2019M10.csv') 
df3.head()
```

```python
df3['date'] = df3['year'].map(str) + '-' + df3['month'].map(str)
df3['date'] = pd.to_datetime(df3['date'])
df3['total arrivals'] = df3.loc[:, 'total arrivals'].apply(lambda x: float(x.replace(",", "")))
df3['total arrivals'].replace(",", "")
df3.head()
```

```python
df4 = df3[['market regions / markets','date','total arrivals','year on year (yoy) growths (%)']]
df4.columns = ['market regions','date','total arrivals','year on year (yoy) growths (%)']
df4.head()
```

```python
df4.to_csv('Tourism_Statistics_Region_2016M10-2019M10(cleaned).csv',index = False)
```
