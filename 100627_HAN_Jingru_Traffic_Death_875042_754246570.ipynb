```python
index = np.arange(5) #Make a Multi-sequence bar chart 
China = [19.2,26.8,8.1,26.1,19.8]
Japan = [32.4,17.4,13.7,36.2,0.3]
Italy = [45.4,25.5,7.4,16.2,5.5]
UK = [46.3,19.3,6.4,22.9,5.2]
USA=[64.4,14.8,2.2,14.1,4.5]
a = 0.15
plt.title('Road traffic deaths Data by country')
plt.xlabel('Country')
plt.ylabel('rate')
#Define the name of the x and y axis
plt.bar(index, 
        China, a, color = 'orange', label = 'china')
plt.bar(index+a, 
        Japan, a, color = 'navajowhite', label = 'Japan')
plt.bar(index+2*a, 
        Italy, a, color = 'tan', alpha = 0.5,label = 'Italy')
plt.bar(index+3*a, 
        UK, a, color = 'chocolate', alpha = 0.5,label = 'US')
plt.bar(index+4*a, 
        USA, a, color = 'sandybrown', alpha = 0.5,label = 'USA')
#Select my favourite color
plt.xticks(index+1.5*a, ['4-wheeled vehicles','motorized 2- or 3-wheelers','Cyclists','Pedestrians','Other'],rotation=90)
#Adjust the slope of x name
plt.legend()
```

### I made a multi-sequence bar chart that figures out the highest traffic death rates in these five countries and compare the proportion of road sections where traffic deaths occur in China and the United States.The analysis found that:Among the five countries (China, Japan, Italy, the United Kingdom and the United States) the country with the highest proportion of traffic deaths in ‘4-wheeled vehicles’ in 2013 was the United States (64.4 percent of the total traffic deaths), while the lowest was China (19.2 percent of the total traffic deaths). In contrast, the highest proportion of deaths in ‘motorized 2- or 3-wheelers’ is China (26.8 percent of the total traffic deaths) and the lowest is United States (14.8percent of the total traffic deaths).The road type with the highest mortality rate in Japan is ‘Pedestrians’. All countries have the least chance of dying from traffic accidents on ‘Cyclists’.

### By making pie charts of road traffic death in five countries, we can draw the conclusion that the proportion in China is more balanced, basically between 27 and 19 per rate. However, 60% of the traffic mortality rate in the United States occurs in ‘4-wheeled vehicles’. Japan is the most prone to accidents on ‘Pedestrians’. The proportion of UK and Italy is quite similar. This reflects that road traffic safety measures in the United States should focus on ‘motorized 2- or 3-wheelers’. Japan needs to strengthen the safety of ‘Pedestrians’ to reduce traffic accidents. UK and Italy, as well as China, need to continue to develop in an all-round way.
